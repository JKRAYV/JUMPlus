```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas
```

```python
initialPD = pandas.read_csv("players_20.csv")

initialPD = initialPD.drop_duplicates()

# List of columns to remove
columns_to_remove = ['player_url', 'real_face', 'player_tags', 'loaned_from', 'joined',
                     'contract_valid_until', 'nation_position', 'nation_jersey_number', 'player_traits']

# Remove the columns
FifaPD = initialPD.drop(columns=columns_to_remove)



# Save the cleaned dataframe to a new CSV file
FifaPD.to_csv('cleaned_file.csv', index=False)
```

```python
FifaPD
```

```python
for x in FifaPD:
    print(f"{x} count: {FifaPD[x].count()}")
```

```python
print(f"This dataset has {len(FifaPD.columns)} columns.")
```

```python
countryCountPD = pandas.DataFrame(FifaPD['nationality'].value_counts())

countryCountPD = countryCountPD.reset_index()
countryCountPD = countryCountPD.rename(columns={'nationality': 'count'})
countryCountPD = countryCountPD.rename(columns={'index': 'nationality'})
```

```python
print("Player Count by nation")
#countryCountPD
countryCountPD
```

```python
print("Top 10 Count by Nation\n")
countryCountPD.sort_values('count').nlargest(10, 'count')
```

```python
#setting style to dark to properly display hue
sns.set(style='dark')
sns.set_style("darkgrid", {"axes.facecolor": ".2"})

#creation of barplot (uses green hue)
plot = sns.barplot(data = countryCountPD.nlargest(5, 'count'), y = 'count', x = 'nationality', color = 'green')
plot.set_xticklabels(plot.get_xticklabels(), rotation=45)
#modifying legend
legend = plt.legend()
legend.axes.set_xlabel('Nation')
legend.axes.set_ylabel('Number of Players')
for text in legend.get_texts():
    text.set_color("white")
```

```python
print("Names and Wages by Top 5 Overall")
top5PlayerWage = FifaPD.nlargest(5, 'overall')[['short_name', 'wage_eur']]
top5PlayerWage
```

```python
#setting style to dark to properly display hue
sns.set(style='dark')
sns.set_style("darkgrid", {"axes.facecolor": ".2"})

#creation of barplot (uses green hue)
plot = sns.barplot(data = top5Wage, y = 'wage_eur', x = 'short_name', color = 'green')
plot.set_xticklabels(plot.get_xticklabels(), rotation=45)
#modifying legend
legend = plt.legend()
legend.axes.set_xlabel('Player')
legend.axes.set_ylabel('Salary in Euros')
for text in legend.get_texts():
    text.set_color("white")
```

```python
print("Top 5 Wages by Player in Germany")
top10WageGermany = FifaPD[FifaPD['nationality'] == 'Germany'].nlargest(10, 'overall')[['short_name', 'overall', 'team_position', 'wage_eur']]
top10WageGermany
```

```python
#Selection of top 5 statitistics from germany
top5HeightGermany = FifaPD[FifaPD['nationality'] == 'Germany'].nlargest(5, ['height_cm'])[['short_name','height_cm', 'weight_kg','wage_eur']]
top5WeightGermany = FifaPD[FifaPD['nationality'] == 'Germany'].nlargest(5, 'weight_kg')[['short_name','height_cm', 'weight_kg','wage_eur']]
top5WageGermany = FifaPD[FifaPD['nationality'] == 'Germany'].nlargest(5, 'wage_eur')[['short_name','height_cm', 'weight_kg','wage_eur']]

#Combining top 5 statistics from germany
top5HeightWeightAgeGermany = top5HeightGermany.merge(top5WeightGermany, on=['short_name', 'height_cm', 'weight_kg', 'wage_eur'], how='outer').merge(top5WageGermany, on=['short_name', 'height_cm', 'weight_kg', 'wage_eur'], how='outer')

print(f"Height, Weight, and Wage by Top 5 Height in Germany\n{top5HeightGermany}\n\n \
Height, Weight, and Wage by Top 5 Weight in Germany\n{top5WeightGermany}\n\n \
Height, Weight, and Wage by Top 5 Wage in Germany\n{top5WageGermany}\n\n \
Top 5 Height, Weight, and Wage  in Germany Combined\n{top5HeightWeightAgeGermany}")



```
